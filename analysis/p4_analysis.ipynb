```python
import re

def load_res(fn):
    with open(fn, "r") as ins:
        scores = []
        state_coverage = []
        for line in ins:
            try:
                arr = re.findall('\d+ \d+.\d+', line )[0].split(' ')
                scores.append(int(arr[0]))
                state_coverage.append(float(arr[1]))
            except:
                pass
    return scores, state_coverage

s1, scov1 = load_res("model1-50binsize.txt")
s2, scov2 = load_res("model2-25binsize.txt")
s3, scov3 = load_res("model3-Mstates_25binsize.txt")
```

```python
%matplotlib inline
import pylab as plt
import numpy as np

plt.rcParams['figure.figsize'] = (14,10)
plt.rcParams['font.size'] = 22

plt.plot(s1, linestyle='None', marker='o', markerfacecolor='r', alpha=1, label='Model1')
plt.plot(s2, linestyle='None', marker='o', markerfacecolor='orange', alpha=1, label='Model2')
plt.plot(s3, linestyle='None', marker='o', markerfacecolor='y', alpha=1, label='Model3')
plt.legend(loc=2)

plt.xlabel('epochs')
plt.ylabel('score')

plt.savefig('scatter_comparison.pdf', dpi=400)
plt.show()

```

```python
n = 100
fs1 = s1[-n:]
fs2 = s2[-n:]
fs3 = s3[-n:]
plt.hist([fs1, fs2, fs3], bins=20, histtype='bar',
         label=['Model1', 'Model2', 'Model3'], color=['r', 'orange', 'y'])
plt.legend()

plt.xlabel('score')
plt.ylabel('count')

plt.savefig('bar_comparison_final' + str(n) + 'epochs.pdf', dpi=400)
plt.show()

print 'model1 max=%d, mean=%f, median=%d' % (np.max(fs1), np.mean(fs1), np.median(fs1))
print 'model2 max=%d, mean=%f, median=%d' % (np.max(fs2), np.mean(fs2), np.median(fs2))
print 'model3 max=%d, mean=%f, median=%d' % (np.max(fs3), np.mean(fs3), np.median(fs3))
```
